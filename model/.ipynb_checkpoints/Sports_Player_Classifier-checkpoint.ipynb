```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("./test_images/kobe_bryant1.jpg",1)
img.shape
```

```python
plt.imshow(img)    
```

```python
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(color_img)
```

```python
gray = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
gray
```

```python
plt.imshow(gray, cmap="gray")
```

```python
face_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_eye.xml")

faces = face_cascade.detectMultiScale(gray, 1.3,5)   #(image, scale_factor, minNeighbors)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
face_img = cv2.rectangle(color_img, (x,y), (x+w, y+h), (255,0,0), 10)
# face_img = cv2.rectangle(image, (upper_left_corner_position),(lower_right_corner_position)
#                         ,(rectangle_color), thickness)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(color_img, (x,y), (x+w, y+h), (255,0,0), 10)
    roi_gray = gray[y:y+h, x:x+w]      #crop region of interest the face in the gray image
    roi_color = face_img[y:y+h, x:x+w]    #get only face in the color img
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for(ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,255,0),5)
plt.figure()
plt.imshow(roi_color)
plt.figure()
plt.imshow(color_img, cmap='gray')
# plt.show()
```

```python
def get_cropped_face_img(image_path):
    color_img = cv2.imread(image_path)
#     color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3,5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = color_img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >=2:
            return roi_color
```

```python
path_to_data = "./dataset/"
path_to_cr_data = "./processed/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)    #append all directories in dataset folder to img_dirs
```

```python
img_dirs
```

```python
import shutil
if os.path.exists(path_to_cr_data):
    shutil.rmtree(path_to_cr_data)     #remove the directory is already exists 
os.mkdir(path_to_cr_data)      #create a "cropped" folder in the given directory
```

```python
cropped_img_dirs = []
sports_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    sports_person = img_dir.split('/')[-1]
    print(sports_person)
    sports_file_names_dict[sports_person] = []
    
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_face_img(entry.path)
        if roi_color is not None:
                cropped_folder = path_to_cr_data + sports_person
                if not os.path.exists(cropped_folder):
                    os.mkdir(cropped_folder)
                    cropped_img_dirs.append(cropped_folder)
                    print("Generating cropped images in folder: ", cropped_folder)
                
                cropped_file_name = sports_person + str(count) + ".png"
                cropped_file_path = cropped_folder + "/" + cropped_file_name
                
                cv2.imwrite(cropped_file_path, roi_color)
                sports_file_names_dict[sports_person].append(cropped_file_path)
                count += 1
print("Processing Finish!")
                    
                    
                    
                    
```
